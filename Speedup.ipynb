```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
###Speedup plot
benchmark = 481
batch = np.array([349.8, 284.3, 270, 254, 244, 228 ])
epoch = np.array([270, 138.7,147.4, 78,84.5,48.3])
plt.plot(np.arange(6), benchmark/batch, label= 'batch')
plt.plot(np.arange(6), benchmark/epoch, label= 'epoch')
plt.legend()
plt.xticks(np.arange(6), ['1worker 2core', '1w4c','2w2c', '2w4c', '4w2c', '4w4c'])
plt.ylabel('Speedup')
plt.xlabel('Infrastructure')
plt.title('Observed Speedup with Different Infrastructure')
```
